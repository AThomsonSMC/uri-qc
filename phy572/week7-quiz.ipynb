### Use Qiskit to display the histogram for the action of a Hadamard gate operating on an input register with 2 qubits each in the |1> state (one Hadamard gate for each qubit).

To create an input register with two 2 qubits in the |1> state, we initialize 2 qubits and operate on each with the X gate. We then apply a Hadamard gate to each to reach the desired state. The qiskit composer produces the following code.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.measure_all()
```

The histogram produced shows that there is an equal probability of measuring all possible states for 2 qubits. That is, there is a 25% chance of measuring |00>, |01>, |10>, and |11> each.

```python
from qiskit.visualization import plot_histogram
from qiskit_aer.primitives import SamplerV2

sampler = SamplerV2()
```

```python
results = sampler.run([circuit], shots=512).result()
plot_histogram(results[0].data.meas.get_counts())
```

This state represents an unentangled superposition of all possible 2-qubit states when measured in the Z basis. This state can also be represented as |-->, and if we measured in the X-axis then we would expect 11 back 100% of the time.
